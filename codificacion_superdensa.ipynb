<h1><center>Codificación Superdensa</center></h1>

La codificación superdensa es un protocolo para enviar información y para esto involucra dos partes: Alice (primera parte) tiene un qubit que está entrelazado con otro qubit el cuál está en manos de Bob (segunda parte). En este proceso, Alice quiere enviar 2 bits de información a Bob y para esto ella sólo realiza operaciones sobre su qubit, lo envía a Bob y éste realiza mediciones sobre dicho qubit que fue enviado y sobre el qubit que ya tenía en posesión.

Uno podría pensar que existe una paradoja, ya que uno de los teoremas de la computación cuántica menciona que de un qubit sólo se puede extraer un bit de información. Entonces ¿cómo es posible enviar dos bits de información transmitiendo un solo qubit? La aparente paradoja se resuelve al considerar que en realidad son dos qubits los que se miden al final, por lo que no hay contradicción con el teorema mencionado. Lo interesante de este protocolo es el hecho de que Alice sólo debe modificar su qubit para poder transmitir los dos bits en la manera que ella decida: 00, 01, 10 o 11. Entonces Alice logra enviar 2 bits de información a Bob mandándole un solo qubit. Alice es capaz de transmitir 2 bits de información interactuando con un solo qubit.

Debemos recordar que un sistema de dos qubits se describe por un vector pertenece a un espacio de Hilbert compuesto $H = H_A \otimes H_B$, en donde $H_A$ es el espacio de Hilbert para el qubit de Alice cuya base ortonormal son los vectores $|0\rangle$ y $|1\rangle$, y $H_b$ es el espacio de Hilbert para el qubit de Bob cuya base ortonormal son los vectores son los vectores $|0\rangle$ y $|1\rangle$.

Supongamos pues que initialmente tenemos dos qubits entrelazados cuyo estado total se describe por uno de los estados de Bell: $$ |\psi \rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)$$ 

en donde el ket de la izquierda (en cada término) representa al qubit de Alice y el ket de la derecha al de Bob. Entonces el primer qubit es enviado a Alice y el segundo a Bob como se muestra en la siguiente figura.



Alice desea enviar dos bits de información, los cuales pueden ser una de las siguientes opciones: 00, 01, 10 o 11. Dependiendo del par de bits a enviar va a ser la operación que Alice deberá aplicar a su respectivo qubit:

<font color=blue>__1.- Par '00'.__</font> Si Alice desea enviar el par '00' entonces no debe modificar su qubit, que matemáticamente se representa como aplicar el operador identidad a su qubit: $$ |\psi_1\rangle = (I \otimes I)|\psi\rangle = (I \otimes I)\frac{1}{\sqrt{2}}(|00\rangle + |11\rangle) = \frac{1}{\sqrt{2}}(I|0\rangle \otimes I|0\rangle + I|1\rangle \otimes I|1\rangle)$$ Por lo que cuando Bob recibe el qubit de Alice el estado de los dos qubits es $$|\psi_1\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)$$

En este punto aún no es viable que Bob realize una medición sobre ambos qubits, ya que el estado de ellos se encuentra tanto en una superposición como en un entrelazamiento, y al realizar una medición obtendremos el par 00 con un 50% de probabilidad y el par 11 con un 50% de probabilidad. Nosotros queremos que el par 00 se mida con 100% de certeza. Para ello Bob primero debe romper ese entrelazamiento entre los dos qubits aplicando una compuerta CNOT en donde el qubit ´control´ es el enviado por Alice y el qubit ´objetivo´ es el que ya estaba en posesión de Bob. Dicho operador lo representamos matemáticamente como $$ CNOT = |0\rangle \langle 0| \otimes I + |1\rangle \langle 1| \otimes \sigma_x $$

En donde $\sigma_x$ (matriz de Pauli en dirección x) actúa como $$ \sigma_x |0\rangle = |1\rangle $$ $$ \sigma_x |1\rangle = |0\rangle $$


Lo que hace este operador es tomar el qubit de Alice, si éste se encuentra en el estado $|0\rangle$ toma al qubit de Bob y no lo modifica. Pero si el qubit de Alice se encuentra en el estado $|1\rangle$ entonces cambia el estado del qubit de Bob:$|0\rangle \longrightarrow |1\rangle$ y $|1\rangle \longrightarrow |0\rangle$.

Por tanto, luego de aplicar esta compuerta tenemos que el estado de los dos qubits es $$ |\psi_2\rangle = (|0\rangle \langle 0| \otimes I + |1\rangle \langle 1| \otimes \sigma_x)(\frac{1}{\sqrt{2}}(|0\rangle \otimes |0\rangle + |1\rangle \otimes |1\rangle)) = \frac{1}{\sqrt{2}}(|0\rangle \otimes |0\rangle + |1\rangle \otimes \sigma_x |1\rangle) =  \frac{1}{\sqrt{2}}(|0\rangle|0\rangle + |1\rangle|0\rangle)$$

Resulta que este estado ya no está entrelazado, ya que se puede escribir como se muestra a continuación $$ |\psi_2\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)\otimes|0\rangle = |+\rangle\otimes|0\rangle$$ En donde $|+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$ es un vector base de $H_A$ en otra dirección, junto con $|-\rangle = \frac{1}{\sqrt{2}}(|0\rangle - |1\rangle)$.

Pero si Bob quiere realizar una medición y obtener el par 00 de Alice, deberíamos tener el estado en términos de la misma base. Otra forma de ver que aún no está listo el estado para ser medido, es que si Bob realiza una medición en el qubit de Alice, tenemos 50% de probabilidad de medir 0 y 50% de medir 1, pero nosotros queremos un 100% de probabilidad de medir 0.

Para ello, Bob aplica una compuerta Hadamard sobre el qubit de Alice. Dicha compuerta opera como sigue: $ H|+\rangle = |0\rangle $ y $ H|-\rangle = |1\rangle $. Por lo que $$ |\psi_3\rangle = (H \otimes I)(|+\rangle \otimes |0\rangle) = H|+\rangle \otimes I|0\rangle = |0\rangle|0\rangle $$

Finalmente, cuando Bob mide el estado $|\psi_3\rangle$ obtiene con 100% de certeza el par de bits 00 que Alice quizo enviar.

<font color=blue>__2.- Par '01'.__</font> Si Alice desea enviar el par '01' entonces debe aplicar la operación $\sigma_x$ anteriormente mencionada sobre su qubit:$$ |\psi_1\rangle = (\sigma_x \otimes I)|\psi\rangle = (\sigma_x \otimes I)\frac{1}{\sqrt{2}}(|00\rangle + |11\rangle) = \frac{1}{\sqrt{2}}(\sigma_x|0\rangle \otimes I|0\rangle + \sigma_x|1\rangle \otimes I|1\rangle)$$ Por lo que cuando Bob recibe el qubit de Alice el estado de los dos qubits es $$|\psi_1\rangle = \frac{1}{\sqrt{2}}(|10\rangle + |01\rangle)$$

Bob aplica la compuerta CNOT: $$|\psi_2\rangle = CNOT|\psi_1\rangle = (|0\rangle \langle 0| \otimes I + |1\rangle \langle 1| \otimes \sigma_x)(\frac{1}{\sqrt{2}}(|1\rangle \otimes |0\rangle + |0\rangle \otimes |1\rangle))  $$

$$ |\psi_2\rangle =\frac{1}{\sqrt{2}} (|0\rangle \otimes |1\rangle + |1\rangle \otimes \sigma_x |0\rangle) = \frac{1}{\sqrt{2}}(|0\rangle|1\rangle + |1\rangle|1\rangle) $$ el cual es un estado no entrelazado que se puede escribir como $$ |\psi_2\rangle = \frac{1}{\sqrt{2}} (|0\rangle + |1\rangle) \otimes |1\rangle = |+\rangle|1\rangle$$

Finalmente Bob aplica una compuerta Hadamard obteniendo el siguiente estado final $$ |\psi_3\rangle = (H \otimes I)(|+\rangle \otimes |1\rangle) = H|+\rangle \otimes I|1\rangle = |0\rangle|1\rangle$$ de tal manera que cuando se mida dicho estado se obtiene el par '01'.

<font color=blue>__3.- Par '10'.__</font> Si Alice desea enviar el par '10' entonces debe aplicar la operación $\sigma_z$ (matriz de Pauli en dirección z) sobre su qubit. Dicho operador acciona de la siguiente manera: $\sigma_z|0\rangle = |0\rangle$ y $\sigma_z|1\rangle = -|1\rangle$. Entonces el estado se convierte en

$$ |\psi_1\rangle = (\sigma_z \otimes I)|\psi\rangle = (\sigma_z \otimes I)\frac{1}{\sqrt{2}}(|00\rangle + |11\rangle) = \frac{1}{\sqrt{2}}(\sigma_z|0\rangle \otimes I|0\rangle + \sigma_z|1\rangle \otimes I|1\rangle)$$ Por lo que cuando Bob recibe el qubit de Alice el estado de los dos qubits es $$|\psi_1\rangle = \frac{1}{\sqrt{2}}(|00\rangle - |11\rangle)$$

Bob aplica la compuerta CNOT: $$|\psi_2\rangle = CNOT|\psi_1\rangle = (|0\rangle \langle 0| \otimes I + |1\rangle \langle 1| \otimes \sigma_x)(\frac{1}{\sqrt{2}}(|0\rangle \otimes |0\rangle - |1\rangle \otimes |1\rangle))$$

$$ |\psi_2\rangle =\frac{1}{\sqrt{2}} (|0\rangle \otimes |0\rangle - |1\rangle \otimes \sigma_x |1\rangle) = \frac{1}{\sqrt{2}}(|0\rangle|0\rangle - |1\rangle|0\rangle) $$ el cual es un estado no entrelazado que se puede escribir como $$ |\psi_2\rangle = \frac{1}{\sqrt{2}} (|0\rangle - |1\rangle) \otimes |0\rangle = |-\rangle|0\rangle$$

Finalmente Bob aplica una compuerta Hadamard obteniendo el siguiente estado final $$ |\psi_3\rangle = (H \otimes I)(|-\rangle \otimes |0\rangle) = H|-\rangle \otimes I|0\rangle = |1\rangle|0\rangle$$ de tal manera que cuando se mida dicho estado se obtiene el par '10'.

<font color=blue>__4.- Par '11'.__</font> Si Alice desea enviar el par '11' entonces debe aplicar la operación $\sigma_z \sigma_x$ sobre su qubit. Entonces el estado se convierte en

$$ |\psi_1\rangle = (\sigma_z\sigma_x \otimes I)|\psi\rangle = (\sigma_z\sigma_x \otimes I)\frac{1}{\sqrt{2}}(|00\rangle + |11\rangle) = \frac{1}{\sqrt{2}}(\sigma_z\sigma_x|0\rangle \otimes I|0\rangle + \sigma_z\sigma_x|1\rangle \otimes I|1\rangle) = \frac{1}{\sqrt{2}}(\sigma_z|1\rangle \otimes |0\rangle + \sigma_z|0\rangle \otimes |1\rangle) = \frac{1}{\sqrt{2}}(-|1\rangle \otimes |0\rangle + |0\rangle \otimes |1\rangle)$$ Por lo que cuando Bob recibe el qubit de Alice el estado de los dos qubits es $$|\psi_1\rangle = \frac{1}{\sqrt{2}}(-|10\rangle + |01\rangle)$$

Bob aplica la compuerta CNOT: $$|\psi_2\rangle = CNOT|\psi_1\rangle = (|0\rangle \langle 0| \otimes I + |1\rangle \langle 1| \otimes \sigma_x)(\frac{1}{\sqrt{2}}(-|1\rangle \otimes |0\rangle + |0\rangle \otimes |1\rangle))$$

$$ |\psi_2\rangle =\frac{1}{\sqrt{2}} (|0\rangle \otimes |1\rangle - |1\rangle \otimes \sigma_x |0\rangle) = \frac{1}{\sqrt{2}}(|0\rangle|1\rangle - |1\rangle|1\rangle) $$ el cual es un estado no entrelazado que se puede escribir como $$ |\psi_2\rangle = \frac{1}{\sqrt{2}} (|0\rangle - |1\rangle) \otimes |1\rangle = |-\rangle|1\rangle$$

Finalmente Bob aplica una compuerta Hadamard obteniendo el siguiente estado final $$ |\psi_3\rangle = (H \otimes I)(|-\rangle \otimes |1\rangle) = H|-\rangle \otimes I|1\rangle = |1\rangle|1\rangle$$ de tal manera que cuando se mida dicho estado se obtiene el par '11'.

$\textbf{Implementación en Qiskit.}$

El primer paso para la implementación es generar el par entrelazado que compartirán Alice y Bob. Podemos pensar como si otra tercera parte (Charlie) generara este entrelazamiento y luego enviara los qubits a Alice y Bob:



Charlie genera este par entrelazado tomando dos qubits que initialmente ambos están en el estado $|0\rangle$, les aplica una compuerta Hadamard y luego una compuerta CNOT: $$ (CNOT)(H \otimes I)|0\rangle |0\rangle = (CNOT)|+\rangle|0\rangle = (CNOT)\frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)|0\rangle = (CNOT)\frac{1}{\sqrt{2}}(|0\rangle|0\rangle + |1\rangle|0\rangle)$$ $$ = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle) $$

El código siguiente considera el qubit q1 (el más a la izquierda) como el qubit de Alice y el qubit q0 (el más a la derecha) como el de Bob, para coincidir con la notación usual que uno encuentra en los libros.

```python
from qiskit import QuantumCircuit, Aer, transpile
from qiskit.visualization import plot_histogram
```

```python
# creamos este par entrelazado de la misma manera
qc = QuantumCircuit(2,2)
qc.h(1) #aplicamos Hadamard en el qubit 1 (qubit de Alice)
qc.cx(1,0) #aplicamos CNOT con el qubit 1 como control y qubit 0 (qubit de Bob) como objetivo
qc.barrier()
```

```python
# creamos una funcion en donde nosotros le indiquemos el par de bits que Alice quiere enviar y modifique el bit de Alice
# de acuerdo a dicho par de bits ingresado

def mensaje(qc,msj):
    if msj[1] == "1": #si el segundo bit en el par (el correspondiente al qubit de Bob) queremos que tenga un 1, le aplicamos X al qubit de Alice
        qc.x(1) 
    if msj[0] == "1": #si el primer bit en el par (el correspondiente al qubit de Alice) queremos que tenga un 1, le aplicamos Z al qubit de Alice
        qc.z(1)
    qc.barrier()
    return qc
```

```python
mensaje(qc,'10') #ingresamos el par que queremos enviar
```

```python
#una vez que Alice aplica dichos operadores y envia el qubit a Bob, este aplica una compuerta CNOT y una Hadamard

qc.cx(1,0)
qc.h(1)
qc.barrier()
```

```python
#Bob mide ambos qubits:
qc.measure([0],[0])
qc.measure([1],[1])

qc.draw('mpl')
```

```python
# corremos el circuito en simulacion
aer_sim = Aer.get_backend('aer_simulator')
result = aer_sim.run(qc).result()
counts = result.get_counts(qc)
plot_histogram(counts)
```

Observamos que siempre obtenemos el par '10' cuando Bob mide ambos qubits. Un punto clave que nos muestra este proceso es la afirmación de que la información es algo físico, y la mecánica cuántica nos provee de habilidades para poder procesar y manipular dicha información.
